### Product analysis

```python
#Category
x = df.groupby('Category')['Product ID'].count().sort_values(ascending = False)
x.plot(kind='pie', autopct='%1.1f%%', center = (4,4), startangle=90)
plt.title = "Category"
plt.ylable = ''
plt.legend(loc = 'upper left')
plt.show()
```

```python
#sub category
x = df.groupby('Sub-Category')['Product ID'].count().sort_values(ascending = False)
x.plot(kind='bar', color = 'cyan',edgecolor = 'magenta')
plt.xlabel('Product Sub-Category')
plt.ylabel('Counts')
plt.title=('Sub-Category')
```

```python
df.groupby('Product ID')['Sales'].count().sort_values(ascending = False).head(15)
```

```python
#region product
x = df.groupby('Region')['Product ID'].count().sort_values(ascending = False)
x
```

```python
x.plot(kind='pie', autopct='%1.1f%%', center = (4,4), startangle=90)
plt.title = "Region"
plt.ylable = ''
plt.legend(loc = 'lower left')
plt.show()
```

```python
#region and category
x = df.groupby('Region')['Category'].value_counts()
x
```

```python
x.plot(kind = 'bar', color = {'orange', 'cyan', 'green'})
plt.xlabel('Category')
plt.ylabel('Counts')
plt.title = 'Region and Category'
```

```python
# year wise order
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
x = df.groupby(df['Order Date'].dt.to_period('Y')).size()
x
```

```python
x.plot(kind = 'line', color = 'red', marker = '*')
plt.xlabel('Year')
plt.ylabel('Orders')
plt.title = 'Year wise order growth'
```

```python
# moth wise order
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
x = df.groupby(df['Order Date'].dt.to_period('M')).size()
x
```

```python
x.plot(kind = 'line', color = 'red', marker = '*')
plt.xlabel('Year')
plt.ylabel('Orders')
plt.title = 'Year wise order growth'
```

```python
#product order
x = df.groupby('Product Name')['Order ID'].count()
x.idxmax(), x.max()
```

```python
y = x.sort_values(ascending = False).head(10)
```

```python
y.plot(kind = 'bar', color = "Cyan", edgecolor = 'red')
plt.xlabel('Products')
plt.ylabel('Counts')
plt.title = 'Top 10 best saling Product'
```

```python
#shiping mode
x = df.groupby('Region')['Ship Mode'].value_counts()
x
```

```python
x.plot(kind = 'bar', color = {'red', 'green', 'blue', 'purple'})
plt.xlabel('region ShipMode')
plt.ylabel('Counts')
plt.title = 'Region and Ship Mmode'
plt.grid()
```

```python
#best saling city
x = df.groupby('City')['Order ID'].count()
#Top 10 best saling city 
y = x.sort_values(ascending = False).head(10)
y
```

```python
y.plot(kind='bar', color = 'cyan',edgecolor = 'magenta')
plt.xlabel('City')
plt.ylabel('Order')
plt.title=('Top 10 best saling city')
```
